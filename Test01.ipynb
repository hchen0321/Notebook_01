```python
import seaborn as sns

df = pd.read_csv('C:/Users/PD/Downloads/DOTstat_low_v20.csv', skiprows=1)

sns.lmplot(x= 'Time', y= 'pH', data=df)
```

```python
# importing libraries 
import random 
import matplotlib.pyplot as plt 
import numpy as np
  
fig = plt.figure() 
  
# defining subplots 
ax1 = fig.add_subplot(231) 
x = df["Time"]
y = df["RQ"]
ax1.plot(x,y)
ax1.set_ylim([0, 1.5])
ax1.set_xlim([0,48])
plt.ylabel('RQ')

ax2 = fig.add_subplot(232) 
y = df["pH"]
ax2.plot(x,y)
ax2.set_ylim([7.0, 7.3])
ax2.set_xlim([0,48])
plt.ylabel('pH')

ax3 = fig.add_subplot(233) 
y = df["OUR"]
ax3.plot(x,y)
plt.xticks(np.arange (0,50,10))
plt.yticks(np.arange (0,120,10))
plt.ylabel('OUR')

ax4= fig.add_subplot(234) 
y = df["DO"]
ax4.plot(x,y)
plt.xticks(np.arange (0,50,10))
plt.yticks(np.arange (0,100,10))
plt.ylabel('DO')

ax5= fig.add_subplot(235) 
y = df["Base volume pumped"]
y1 = df["Feed#2 volume pumped"]
ax5.plot(x,y)
ax5.plot(x,y1)
plt.xticks(np.arange (0,60,10))
plt.yticks(np.arange (0,20,5))
plt.ylabel('Base/Feed#2 vol pumped, mLs')

plt.show() 
```
